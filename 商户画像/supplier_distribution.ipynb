# 活跃天数

```python
plt.plot(supplier_data.ix[:,'active_day'].value_counts(),'go')
plt.show()
```

```python
supplier_data.ix[:,'active_day'].value_counts().plot(kind='bar',title= 'active_day')
```

```python
supplier_data[supplier_data.city_id == 19].ix[:,'active_day'].value_counts().plot(kind='bar',title= 'active_day')
```

```python
supplier_data[supplier_data.city_id == 1].ix[:,'active_day'].value_counts().plot(kind='bar',title= 'active_day')
```

```python
city_active_day = pd.crosstab(supplier_data.active_day,supplier_data.city_id)
city_active_day.ix[:,2:3].plot(kind = 'bar')
```

```python
city_active_day.ix[23:,[1,2,3,19]].plot(kind = 'bar')
```
